```python
def search_bre_names_nontrad(input_data):
    final_nt_mapping = pd.DataFrame(columns = ['Type', 'matches', 'Class_x'])

    #Search based on the list from the 'Non-traditional' tab in https://docs.google.com/spreadsheets/d/1E-UFBDB_J1fGJgQuEqjvJT9bJvPGI0TG/edit#gid=713873780
    data = input_data.copy()
    data = pd.DataFrame( columns = ['Type'], data = fres.Type.unique().tolist() )
    data = data[~data['Type'].isna()]

    s = ntmapping['Name_x'].tolist()

    m = data['Type'].apply(lambda x: process.extract(x, s, limit = 1))
    data['matches' ] = m

    m2 = data['matches'].apply(lambda x: ', '.join([i[0] for i in x if i[1] >= 90]))
    data['matches' ] = m2

    t = pd.merge(data, ntmapping[['Class_x', 'Name_x']].drop_duplicates(), left_on ='matches', right_on ='Name_x' , how='left') 
    t = t[~t['Name_x'].isna()]

    final_nt_mapping = pd.concat([final_nt_mapping, t])
    
    
    #Now try with the 'Alternate names' tab  
    rows = data[data['matches']==''].index

    d = data.iloc[rows].copy()

    s = ntmapping['Name_y'].tolist()

    m = d['Type'].apply(lambda x: process.extract(x, s, limit = 1))
    d['matches' ] = m

    m2 = d['matches'].apply(lambda x: ', '.join([i[0] for i in x if i[1] >= 90]))
    d['matches' ] = m2


    #Update hawksley to METAL for aluminium bungalows

    t = pd.merge(d, ntmapping[['Class_x', 'Name_y']].drop_duplicates(), left_on ='matches', right_on ='Name_y' , how='left') 
    hwk = t[t['matches']=='Hawksley'].index

    for x in hwk:
        t.at[x,'Class_x']='MET'

    final_nt_mapping = pd.concat([final_nt_mapping, t])
    return final_nt_mapping
```
