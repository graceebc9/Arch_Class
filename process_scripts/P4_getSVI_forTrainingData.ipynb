```python
# load the geotasgged address
gdf = pd.read_csv('geocoded_addresses1.csv')
gdf.drop(columns='Unnamed: 0', inplace=True)

import hashlib
import pandas as pd

# concatenate the values in the relevant columns into a single string - replicable
gdf['id_string'] =  gdf['Address'] + gdf['raw_Construction_type'] + gdf['BRE_Class'] + gdf['Latitude'].astype(str) + gdf['Longitude'].astype(str)

# hash the concatenated string using the SHA-256 algorithm to generate a unique ID
gdf['unique_id'] = gdf['id_string'].apply(lambda x: hashlib.sha256(x.encode('utf-8')).hexdigest())

# view the resulting dataframe with the new unique_id column
gdf.head()
```
